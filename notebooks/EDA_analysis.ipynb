```python
import pandas as pd

# Load the Apple historical data
aapl_data = pd.read_csv('../data/yfinance_data/AAPL_historical_data.csv')
```

```python
aapl_data.describe()
aapl_data.head()
aapl_data.info()
```

```python
aapl_data['Date'] = pd.to_datetime(aapl_data['Date'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(aapl_data['Date'], aapl_data['Close'], label='Close Price', color='blue')
plt.title('AAPL Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(aapl_data['Date'], aapl_data['Volume'], label='Volume', color='orange')
plt.title('AAPL Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.show()
```

```python
aapl_data['Daily Return'] = aapl_data['Adj Close'].pct_change()
```
